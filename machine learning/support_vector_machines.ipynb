```python
from sklearn.svm import SVC
```

```python
# help(SVC)#c allwow missclassifation inversly proportion to cin the equate
```

```python

y = df['Virus Present']
X = df.drop('Virus Present',axis=1) 
```

```python
model = SVC(kernel='linear',C=1000 )
```

```python
model.fit(X,y)
```

```python
# This is imported from the supplemental .py file
# https://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
#classification with scikit learn 

model=SVC(kernel='linear',C=0.05)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)#c is inversly proportional to soft margin
```

```python
model=SVC(kernel='rbf',C=1,gamma='scale')
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
# model=SVC(kernel='rbf',C=1,gamma='auto')# gamma defines a single training example have
# model.fit(X,y)#more nois form support vectors
# plot_svm_boundary(model,X,y)
```

```python
#if gama is large like 2 it will overfit hear to much nois
```

```python
model=SVC(kernel='sigmoid')# gamma defines a single training example have
model.fit(X,y)#more nois form support vectors
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='poly',C=0.05,degree=4)# gamma defines a single training example have
model.fit(X,y)#more nois form support vectors
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python


svm = SVC()
param_grid = {'C': [0.01, 0.1, 1], 'kernel': ['linear', 'rbf']}
```

```python
grid= GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```
