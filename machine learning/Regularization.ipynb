```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
filepath = r'C:\Users\parim\Downloads\Advertising (1).csv' 
df = pd.read_csv(filepath)
```

```python
df = df.drop('ID',axis=1)
```

```python
df.head()
```

```python
X=df.drop('Sales',axis=1)
```

```python
y=df['Sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter=PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features=polynomial_converter.fit_transform(X)
```

```python
X.shape
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()
```

```python
#we do not want information leaking from test state

scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train[0]
```

```python
poly_features
```

```python
from sklearn.linear_model import Ridge
```

```python
# help(Ridge)
```

```python
ridge_model=Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_prediction=ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE=mean_absolute_error(y_test,test_prediction)
```

```python
MAE
```

```python
RMSE=np.sqrt(mean_squared_error(y_test,test_prediction))
```

```python
RMSE
```

```python
from sklearn.linear_model import RidgeCV#perform crossvalidations for alpha value ,best alpha value perfomance based of cross validation,cross validation - average error metric for all of these
```

```python
ridge_cv_model=RidgeCV(alphas=(0.1,1.0,10.0),scoring='neg_mean_absolute_error')#how many folds we are going to do in cross validation
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```
