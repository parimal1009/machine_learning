```python
a =np.linspace(0,10,11)

b= a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
axes[0][1].plot(x,y)
axes[1][0].plot(a,b)

plt.tight_layout()


```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(4,10),dpi=200)

axes[0, 0].plot(x, y)
axes[1, 1].plot(a, b)
axes[0, 1].plot(x, y)
axes[1, 0].plot(a, b)
axes[1, 0].set_ylabel('y label 1,0')

fig.subplots_adjust(wspace=0.9, hspace=0.5) # Adjust width and height

fig.suptitle("figure level",fontsize=16)

plt.show()

fig.savefig('new_subplots.png',bbox_inches='tight')
```
