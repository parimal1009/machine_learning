```python
import pandas as pd

# Corrected file path with raw string notation
filepath = r'C:\Users\parim\Downloads\StudentsPerformance.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(filepath)

# Now `df` contains the data from StudentsPerformance.csv
```

```python
df.head()
```

```python
sns.boxplot(data=df, y='reading score', x='parental level of education',hue='test preparation course',palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.violinplot(data=df, y='reading score', x='parental level of education',hue='test preparation course',palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.violinplot(data=df, y='reading score', x='parental level of education',hue='test preparation course',palette='Set2',inner='quartile)'
               )

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course',dodge=True)#dodge splits the swarmplot
plt.legend(bbox_to_anchor=(1.35, 0.5))
```
