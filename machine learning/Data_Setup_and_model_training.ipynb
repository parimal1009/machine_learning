```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
filepath = r'C:\Users\parim\Downloads\Advertising (1).csv'
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))

axes[0].plot(df['TV'], df['Sales'], 'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['Radio'], df['Sales'], 'o')
axes[1].set_title("Radio Spend")

axes[2].plot(df['Newspaper'], df['Sales'], 'o')
axes[2].set_title("Newspaper Spend")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df)
```

```python
X=df.drop('Sales',axis=1)
```

```python
X
```

```python
y=df['Sales']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(df)
```

```python
X_train
```

```python
len(X_test)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# model.predict(X_test)#y-test
```

```python
# X_test.head()
```

```python
# y_test
```

```python
test_predictions = model.predict(X_test)
```

```python
# test_predictions
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['Sales'].mean()
```

```python
sns.histplot(data=df,x='Sales',bins=20)
```

```python
mean_absolute_error(y_test,test_predictions)
```

```python
#root mean squared error standard deviation
np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
#mean squared error variants

mean_squared_error(y_test,test_predictions)
```

```python
test_residuals = y_test-test_predictions
```

```python
test_residuals
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='r',ls='--')
```

```python
sns.displot(test_residuals,bins=25,kde=True)
```

```python
import scipy as sp  
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 8), dpi=100)

# Generate a Q-Q plot of the test_residuals
_ = stats.probplot(test_residuals, plot=ax)

# Show the plot
plt.show()
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
final_model.coef_#b coefficients of tv,radio,newsoaoer
```

```python
X.head()
```

```python
y_hat = final_model.predict(X)
```

```python
#true vs predicted values
#points in blue are true values
#points in red are predicted values
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))

axes[0].plot(df['TV'], df['Sales'], 'o')
axes[0].plot(df['TV'],y_hat,'o',color='red')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['Radio'], df['Sales'], 'o')
axes[1].plot(df['Radio'],y_hat,'o',color='red')
axes[1].set_title("Radio Spend")

axes[2].plot(df['Newspaper'], df['Sales'], 'o')
axes[2].plot(df['Newspaper'],y_hat,'o',color='red')
axes[2].set_title("Newspaper Spend")

plt.tight_layout()
plt.show()
```
